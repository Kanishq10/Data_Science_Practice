```python
import pandas as pd
data=pd.DataFrame([75, 97, 70, 25, 80, 40, 82, 72, 82, 46, 72, 64, 24, 66, 78, 34, 37,
    64, 75, 94, 76, 96, 88, 37, 29, 51, 67, 45, 69, 59, 81, 66, 30, 52,
    66, 85, 80, 74, 69, 83, 66, 84, 76, 73, 77, 85, 72, 71, 71, 75, 59,
    84, 65, 64, 86, 58, 85, 61, 79, 61, 68, 61, 60, 78, 76, 80, 57, 75])

data.describe()
```

```python
#Detect Outliars
#detecting with 68-95-99 rule
upper_bound=data.mean() + 3 * data.std()
lower_bound=data.mean() + 3 * data.std()

print(upper_bound,lower_bound)
#values outside these are outliars
```

```python
#with iqr
Q3=data.quantile(0.75)
Q1=data.quantile(0.25)
IQR=Q3-Q1
upper=Q3+(1.5*(IQR))
lower=Q1-(1.5*(IQR))
upper,lower
```

```python
outliars=data[data>upper] & data[data<lower]
len(outliars)
```

```python
data=data[(data<=upper) & (data>=lower)]    #removing outliar
data
```

### Outliar: value beyond the good range

For detecting outliars:<br>
#### In case of normal distribuation - we follow the imperical rule (65-95-99.7)
- In particular, the empirical rule predicts that in normal distributions, 68% of observations fall within the first standard deviation (Âµ Â± Ïƒ), 95% within the first two standard deviations (Âµ Â± 2Ïƒ), and 99.7% within the first three standard deviations (Âµ Â± 3Ïƒ) of the mean.



#### In case of skew distribution - we follow IQR (inter quantile range)
IQR= q3-q1
- lower bound = Q1-1.5*IQR
- upper bound = Q3+1.5*IQR
- range=(lower value -> upper value)
- value beyond this range is outliar
<br><br>
- lower bound = Q1-3*IQR
- upper bound = Q3+3*IQR
- range=(lower value -> upper value)
- value beyond this range is extreme outliar <br><br>

Decile: it divides the datasets in 10 equal parts:
<br> The first decile (D1) divides the data such that 10% of the data is below D1 and 90% above it.<br>
The fifth decile (D5) divides the data into two equal parts:
- ie. D5=median=Q2
