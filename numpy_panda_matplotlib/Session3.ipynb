```python
import pandas as pd
import numpy as np
import os

#to ignore warnings
import warnings
warnings.filterwarnings('ignore')

files=os.listdir('/Users/kanis/Documents/Sales_Data')
files
```

```python
#Read the files
df = pd.DataFrame()

for i in files:
    data=pd.read_csv(os.path.join('/Users/kanis/Documents/Sales_Data',i))
    df=pd.concat([df,data])

df
```

```python
#as null values are same and it means there exist a whole null row, then we drop all null values
df.dropna(inplace=True)
df.shape
```

#### Changing Data Type

```python
#getting unique value for each category
df[['Order ID','Product','Quantity Ordered','Price Each','Order Date','Purchase Address']].describe()
```

```python
df['Quantity Ordered'].unique()
```

```python
#create the filter to drop the text value from quantity ordered
filter_text= df['Quantity Ordered'] != 'Quantity Ordered'

df=df[filter_text]

df['Quantity Ordered'].unique()
```

```python
#convert the quantity ordered and price each data type
df['Quantity Ordered']=df['Quantity Ordered'].astype('float')
df['Price Each'] = df['Price Each'].astype('float')
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df.dtypes
```

```python
#create the Year, Month , Hour , Minute columns
df['Year']= df['Order Date'].dt.year
df['Month']= df['Order Date'].dt.month
df['Hour']= df['Order Date'].dt.hour
df['Minute']= df['Order Date'].dt.minute
```

```python
#create the revenue column
df['Revenue']=df['Price Each'] * df['Quantity Ordered']

m='India, vs, SriLanka'
m.split(',')[1]
```

```python
city=[]
for i in df['Purchase Address']:
    c=i.split(',')[1]
    city.append(c)

df['City']=city
df
```
