```python
iris=pd.read_csv('resources/iris.csv') 
```

```python
years=[2015,2016,2017,2018,2019,2020,2021]
tesla=[320,335,360,350,375,380,400]
apple=[220,449,322,444,432,500,501]
```

```python
plt.figure(figsize=(16,10))  # creates a figure object with the specified width and height in inches.
plt.plot(years,tesla,marker='o')
plt.plot(years,apple, marker='*',markersize=10, markerfacecolor='blue')
plt.title("Comparison b/w two stocks")
plt.xlabel('Years')
plt.ylabel('Value of the Stock per year')
plt.legend(['Tesla','Apple'])
plt.show()
```

```python
maximum=iris['petal length in cm'].max()
iris['petal length in cm']=iris['petal length in cm'].replace(maximum,7.2)
```

```python
minimum=iris['petal length in cm'].min()
iris['petal length in cm']=iris['petal length in cm'].replace(minimum,1.2)
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=iris['petal length in cm'],y=iris['petal width in cm'],hue=iris['outputs'])
```
