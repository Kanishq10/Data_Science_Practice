Join

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Id':[1,2,3,4],'Name':['Mia','Sara','Joey','Karan'],'Age':[47,38,48,41],'Gender':['Male','Female','Male','Male']})
df
```

```python
df.rename(columns={'Gender':'Sex'})
```

```python
bf= pd.DataFrame({'Emp_Id':[101,102,103,104,105,106,107],'Name':['Mia','Lana','Sara','Babu','Joey','Karan','Jhonny'],'Salary':[4700,3800,48000,4100,4000,7000,2000],'City':['Boston','New York','Berlin','Boston','New York','Berlin','New Jersey']})
bf
```

```python
pd.merge(df,bf,on='Name',how='inner')
```

```python
pd.merge(df,bf,on='Name',how='outer')
```

```python
pd.merge(df,bf,on='Name',how='left')
```

```python
pd.merge(df,bf,on='Name',how='right')
```
