### Outliar: value beyond the good range

For detecting outliars:<br>
#### In case of normal distribuation - we follow the imperical rule (65-95-99.7)
- mu +- 3sigma are called outliars in normal

![WhatsApp Image 2023-08-25 at 14.44.21.jpg](<attachment:WhatsApp Image 2023-08-25 at 14.44.21.jpg>)

#### In case of skew distribution - we follow IQR (inter quantile range)
IQR= q3-q1
- lower value = xi-1.5*IQR
- upper bound = xi+1.5*IQR
- range=(lower value -> upper value)
- value beyond this range is outliar

- lower value = xi-3*IQR
- upper bound = xi+3*IQR
- range=(lower value -> upper value)
- value beyond this range is extreme outliar
